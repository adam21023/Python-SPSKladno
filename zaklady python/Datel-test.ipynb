Vytvořte třídu `Kniha` s těmito atributy (nastavte v `__init__`): **(1b)**

* `nazev`
* `autor`
* `zanr`
* `pocet_stran` — ověřte, že hodnota je `int` (pokud ne, vyvolejte `TypeError`)

Dále implementujte: **(1b)**
* pro `pocet_stran` vytvořte `@property` s `@pocet_stran.setter`, který kontroluje typ
* metodu `__str__`, která vrátí info o knize v čitelném formátu

````python
# TODO class Kniha
class Kniha:
    def __init__(self, nazev, autor, zanr, pocet_stran):
        self.nazev = nazev
        self.autor = autor
        self.zanr = zanr
        self.pocet_stran = pocet_stran 

    @property
    def pocet_stran(self):
        return self._pocet_stran

    @pocet_stran.setter
    def pocet_stran(self, hodnota):
        if not isinstance(hodnota, int):
            raise TypeError("pocet_stran musí být typu int")
        self._pocet_stran = hodnota

    def __str__(self):
        return f"Kniha : {self.nazev}, Autor: {self.autor}, žánr: {self.zanr}, počet stran: {self.pocet_stran}"
        
````

# 2
Vytvořte třídu `Knihovna` s těmito atributy (nastavte v `__init__`): **(1b)**
* `nazev`
* `knihy` — seznam knih (výchozí prázdný seznam)
* `kapacita` — maximální počet knih

Implementujte tyto metody: **(1b + 2b)**

* `__str__` — vrátí přehled o knihovně (název, obsazenost, výpis knih)
* `__len__` — vrátí aktuální počet knih
* `__contains__(self, kniha)` — umožní použití operátoru `in` (porovnání podle názvu)
* `pridej_knihu(self, kniha)` a `odeber_knihu(self, kniha)` — přidání/odebrání knihy s kontrolou kapacity a duplicit + zpětná vazba pomocí `print()` **(+2b)**

````python
#TODO class Knihovna
class Knihovna:
    def __init__(self, nazev, kapacita):
        self.nazev = nazev
        self.kapacita = kapacita
        self.knihy = []

    def __str__(self):
        vypis = ""
        for kniha in self.knihy:
            vypis += str(kniha) + "\n"

        return f"{self.nazev} ({len(self)}/{self.kapacita})\n{vypis}"

    def __len__(self):
        return len(self.knihy)

    def __contains__(self, kniha):
        for k in self.knihy:
            if k.nazev == kniha.nazev:
                return True
        return False

    def pridej_knihu(self, kniha):
        if len(self.knihy) >= self.kapacita:
            print("Knihovna je plná.")
        elif kniha in self:
            print("Kniha už v knihovně je.")
        else:
            self.knihy.append(kniha)
            print("Kniha přidána.")

    def odeber_knihu(self, kniha):
        if kniha in self:
            self.knihy = [k for k in self.knihy if k.nazev != kniha.nazev]
            print("Kniha odebrána.")
        else:
            print("Kniha v knihovně není.")
````
